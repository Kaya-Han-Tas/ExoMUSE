```python
df_300 = pd.read_csv("../../notebooks/toi2431-analysis/df_planet_300_2sigma.csv")
df_300
```

```python
def get_vals(vec):
    fvec   = np.sort(vec)

    fval  = np.median(fvec)
    nn = int(np.around(len(fvec)*0.15865))

    vali,valf = fval - fvec[nn],fvec[-nn] - fval
    return fval,vali,valf
```

```python
for i in range(len(df_300)):
    dataset = juliet.load(input_folder = f'../../notebooks/toi2431-analysis/300 Planet Injection Run Results/toi_2431_fit_{i}', 
                          out_folder = f'../../notebooks/toi2431-analysis/300 Planet Injection Run Results/toi_2431_fit_{i}')
    results = dataset.fit(sampler = "dynamic_dynesty")

    K_fit, lower_err, upper_err = get_vals(results.posteriors['posterior_samples']['K_p1'])
    K_truth = df_300["K_truth (m/s)"].iloc[i]

    df_300.loc[i, "K_fit Lower Error"] = lower_err
    df_300.loc[i, "K_fit Upper Error"] = upper_err
```
