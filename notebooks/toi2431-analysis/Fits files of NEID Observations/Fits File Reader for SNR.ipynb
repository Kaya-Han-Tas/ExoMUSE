```python
from astropy.io import fits
```

```python
#We import the fits files.
fits1 = "neidL2_20241217T043034.fits"
fits2 = "neidL2_20241217T075226.fits"
fits3 = "neidL2_20241219T072948.fits"
fits4 = "neidL2_20241220T030503.fits"
fits5 = "neidL2_20241221T015204.fits"
fits6 = "neidL2_20250105T032137.fits"
fits7 = "neidL2_20250105T054814.fits"
fits8 = "neidL2_20250110T044504.fits"
fits9 = "neidL2_20250219T024820.fits"
fits10 = "neidL2_20250219T025347.fits"
fits11 = "neidL2_20250219T040733.fits"
fits12 = "neidL2_20250219T041301.fits"
```

```python
#We define a function that takes the file, reads the SNR and returns it.
def fits_snr_reader(filename):
    #We open the file.
    with fits.open(filename) as hdul:
        #We print out the header of the file.
        print(hdul[0].header)

        #We extract the value of SNR if it exists.
        SNR_value = hdul[0].header.get("EXTSNR", "Key not found")
        print(f"SNR Value: {SNR_value}")
```

```python
#We print out the SNRs.
fits_snr_reader(fits1)
fits_snr_reader(fits2)
fits_snr_reader(fits3)
fits_snr_reader(fits4)
fits_snr_reader(fits5)
fits_snr_reader(fits6)
fits_snr_reader(fits7)
fits_snr_reader(fits8)
fits_snr_reader(fits9)
fits_snr_reader(fits10)
fits_snr_reader(fits11)
fits_snr_reader(fits12)
```
