```python
import ironman
import numpy as np
import pandas as pd
import corner
import matplotlib.pyplot as plt
import rmfit
from astropy import units as u
from astropy import constants as c
import juliet
from astropy.timeseries import LombScargle
```

```python
df_300 = pd.read_csv("df_planet_300_2sigma.csv")
df_300
```

```python
for i in range(len(df_300)):
    dataset = juliet.load(input_folder = f'../../notebooks/toi2431-analysis/300 Planet Injection Run Results/toi_2431_fit_{i}', 
                          out_folder = f'../../notebooks/toi2431-analysis/300 Planet Injection Run Results/toi_2431_fit_{i}')
    results = dataset.fit(sampler = "dynamic_dynesty")

    K_fit = results.posteriors['posterior_samples']['K_p1']
    K_truth = df_300["K_truth (m/s)"].iloc[i]

    if (K_truth > np.percentile(K_fit, 0.15)) and \
       (K_truth < np.percentile(K_fit, 99.85)):
        df_300.loc[i, "Result"] = "Recovered"
    
    else:
        df_300.loc[i, "Result"] = "Not Recovered"
```

```python
df_300
```

```python
print(f"Number of Recovered Injections: {df_300[df_300['Result'] == 'Recovered'].shape[0]}")
print(f"Number of Not Recovered Injections: {df_300[df_300['Result'] == 'Not Recovered'].shape[0]}")
```

```python
df_300.to_csv("df_planet_300_3sigma.csv", index=False)
```
